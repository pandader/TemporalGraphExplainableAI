```python
import os
import warnings
import numpy as np
import pandas as pd
import networkx as nx
import stella
warnings.simplefilter('ignore')
```

```python
### load title name entity tags and short hand token
IMPROT_FILE = 'validData_Reduced.csv'
STANDARD_TITLE_FILES = ['tech_jobs.csv', 'non_tech_jobs.csv']
IMPORT_PATH = '/Users/nirvana/Dropbox/Repository/Production/TitleGroupingProj/data/utilityData/' if os.name == 'posix' \
    else r'C:\Users\Jay\Dropbox\Repository\Production\TitleGroupingProj\data\utilityData'
```

```python
### import golden source
# load standard titles
df_titles = pd.concat([
    pd.read_csv(os.path.join(IMPORT_PATH, STANDARD_TITLE_FILES[0])), 
    pd.read_csv(os.path.join(IMPORT_PATH, STANDARD_TITLE_FILES[1]))]).drop(columns=['Edu'])
df_titles.GROUP = df_titles.GROUP.str.lower()
df_titles.TITLE = df_titles.TITLE.str.lower()
# load cleaned profiles
df = pd.read_csv(os.path.join(IMPORT_PATH, IMPROT_FILE))
df['Start_Year'] = df['Start_Date'].apply(lambda x: int(x.split('-')[0]))
df['End_Year'] = df['End_Date'].apply(lambda x: int(x.split('-')[0]))
org_employee_id_2_new_id = {id : idx + 1 for idx, id in enumerate(df.ID.unique())}
df['EmployeeID'] = df.ID.apply(lambda x: org_employee_id_2_new_id[x])
org_title_2_new_id = {id : idx + 1 for idx, id in enumerate(df.Title.unique())}
df['TitleID'] = df.Title.apply(lambda x: org_title_2_new_id[x])
df.head(3)
```

```python
### pick a window
window_s, window_e = 2016, 2018
cond1 = (df.End_Year >= window_s) & (df.End_Year <= window_e)
cond2 = (df.Start_Year >= window_s) & (df.Start_Year <= window_e)
df_window = df[cond1|cond2].reset_index(drop=True)
## sanity check code
# idx = 10
# id = df_window.ID.unique()[idx]
# df_window[df_window.ID == id]
df_window.head(3)
```

```python
### build bipartite graph (ET)
tmp_df_graph = df_window[['EmployeeID', 'TitleID']]
nx_g = nx.DiGraph()
for each in tmp_df_graph.EmployeeID: nx_g.add_node(each, label="E") # employee
for each in tmp_df_graph.TitleID: nx_g.add_node(each, label="T") # title
for _, row in tmp_df_graph.iterrows():
    nx_g.add_edge(row.EmployeeID, row.TitleID)
```
